```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Electric_Vehicle_Population_Data.csv")
df
```

### 1)Distribution of electric vehicle types

```python
type_counts = df['Electric Vehicle Type'].value_counts()
print(type_counts)
plt.figure(figsize=(10,6))
colors=['salmon','orange']
type_counts.plot(kind='bar',color=colors,width=0.4)
plt.xlabel('Electric Vehicle Type')
plt.ylabel('Count')
plt.title('Distribution of Electric Vehicle Types')
plt.show()
```

### 2)Variation of electric range across different vehicle types


```python
range_by_type = df.groupby('Electric Vehicle Type')['Electric Range'].mean()
print(range_by_type)
plt.figure(figsize=(10,6))
colors=['pink','brown']
range_by_type.plot(kind='bar',color=colors,width=0.4)
plt.xlabel('Electric Vehicle Type')
plt.ylabel('Average Electric Range (miles)')
plt.title('Average Electric Range by Vehicle Type')
plt.show()
```

### 3)Top 5 Counties with the Highest Electric Vehicle Population

```python
top_counties = df['County'].value_counts().head(5)
print(top_counties)
plt.figure(figsize=(10,6))
colors=['pink','brown','orange','salmon','blue']
top_counties.plot(kind='bar', color=colors,width=0.4)
plt.title('Top 5 Counties with Highest Electric Vehicle Population')
plt.xlabel('County')
plt.ylabel('Number of Electric Vehicles')
plt.xticks(rotation=45)
plt.show()
```

### 4)Top 10 manufacturers with the highest electric range

```python

top_10_manufacturers = df.groupby('Make')['Electric Range'].max().nlargest(10)
print("Top 10 manufacturers with the highest electric range:")
print(top_10_manufacturers)
plt.figure(figsize=(10, 6))
top_10_manufacturers.plot(kind='bar', color='skyblue')
plt.title('Top 10 Manufacturers with Highest Electric Range')
plt.xlabel('Manufacturer')
plt.ylabel('Electric Range (miles)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### 5)find out Electric Vehicle Sales by Year

```python

sales_by_year = df.groupby('Model Year').size()
sales_by_year = sales_by_year.sort_values(ascending=False)
print(sales_by_year)
# Plotting
plt.figure(figsize=(10, 6))
sales_by_year.plot(kind='line', marker='o', color='green')
plt.title('Electric Vehicle Sales by Year')
plt.xlabel('Year')
plt.ylabel('Number of Electric Vehicles Sold')
plt.grid(True)
plt.tight_layout()
plt.show()

```

### 6)find out Top 10 Highest-Selling Electric Vehicle Models

```python
sales_by_model = df['Model'].value_counts()
top_10_models = sales_by_model.head(10)
print(top_10_models)
plt.figure(figsize=(10, 6))
plt.bar(top_10_models.index,top_10_models.values, color=['red','green', 'lightgreen', 'salmon', 'gold', 'lightcoral', 'lightblue','brown','blue','purple','pink','brown'],label=top_10_models.index)
plt.title('Top 10 Highest-Selling Electric Vehicle Models')
plt.xlabel('Model')
plt.ylabel('Number of Sales')
plt.xticks(rotation=45, ha='right')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

### 7)What are the unique makes present in battery electric vehicles (BEVs), and how many different makes are there in total?

```python
bev_df = df[df['Electric Vehicle Type'] == 'Battery Electric Vehicle (BEV)']

# Get the unique makes in battery electric vehicles
Count_battery_ev = bev_df['Make'].nunique()
battery_ev_makes = bev_df['Make'].unique()

print("Makes in Battery Electric Vehicles:")
print(Count_battery_ev)
print(battery_ev_makes)
```

### 8)What is the maximum electrical range of each Tesla model, and how does it vary across different Tesla models?

```python
tesla_df = df[df['Make'] == 'TESLA']
model_wise_range = tesla_df.groupby('Model')['Electric Range'].max()

print("Tesla Model-Wise Electrical Range:")
print(model_wise_range)
plt.figure(figsize=(10, 6))
colors=['red','green', 'lightgreen', 'salmon', 'gold']
model_wise_range.plot(kind='bar', color=colors)
plt.title('Tesla Model-Wise Electrical Range')
plt.xlabel('Model')
plt.ylabel('Maximum Electrical Range (miles)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### 9)Total Tesla Vehicles by Model in 2020

```python
df['Model Year'] = pd.to_datetime(df['Model Year'], format='%Y')
tesla_df = df[df['Make'] == 'TESLA']
tesla_2020_df = tesla_df[tesla_df['Model Year'].dt.year == 2020]
tesla_2020_total_vehicles_by_model = tesla_2020_df['Model'].value_counts()

print("Tesla Vehicles by Model in 2020:")
print(tesla_2020_total_vehicles_by_model)
plt.figure(figsize=(10, 6))
colors=['green', 'lightgreen', 'salmon', 'gold']
tesla_2020_total_vehicles_by_model.plot(kind='bar', color=colors)
plt.title('Total Tesla Vehicles by Model in 2020')
plt.xlabel('Model')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### 10)Top 10 Cities with the Highest Number of Electric Vehicle Sales

```python
total_sales_by_city = df['City'].value_counts()
top_10_cities = total_sales_by_city.nlargest(10)
print(top_10_cities)

colors = sns.color_palette('pastel')[:len(top_10_cities)]
plt.figure(figsize=(10, 6))
top_10_cities.plot(kind='bar', color=colors)
plt.title('Top 10 Cities with the Highest Number of Electric Vehicle Sales')
plt.xlabel('City')
plt.ylabel('Number of Electric Vehicle Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 11)What is the distribution of Clean Alternative Fuel Vehicle (CAFV) eligibility among Tesla vehicles?

```python
tesla_cafv_counts = df[df['Make'] == 'TESLA']['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].value_counts()
print(tesla_cafv_counts)
tesla_cafv_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightgreen'], startangle=140)
plt.title('Tesla Vehicles by CAFV Eligibility')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

### 12)top 5 BMW models with the highest electrical range

```python

bmw_df = df[df['Make'] == 'BMW']
bmw_model_range = bmw_df.groupby('Model')['Electric Range'].max()


top_5_bmw_models = bmw_model_range.nlargest(5)
print(top_5_bmw_models)

plt.figure(figsize=(10, 6))
top_5_bmw_models.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top 5 BMW Models with the Highest Electrical Range')
plt.xlabel('BMW Model')
plt.ylabel('Maximum Electric Range (miles)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

### 13)Analyze BMW's year-wise sales

```python
df['Model Year'] = pd.to_datetime(df['Model Year'], format='%Y')

bmw_df = df[df['Make'].str.upper() == 'BMW']

bmw_sales_by_year = bmw_df['Model Year'].dt.year.value_counts().sort_index()
print(bmw_sales_by_year)
plt.figure(figsize=(10, 6))
bmw_sales_by_year.plot(kind='line', color='blue', marker='o', linestyle='-', linewidth=2, markersize=6)
plt.title('BMW Year-wise Vehicle Sales')
plt.xlabel('Year')
plt.ylabel('Number of Vehicles Sold')
plt.grid(True)
plt.xticks(bmw_sales_by_year.index, rotation=45)
plt.tight_layout()
plt.show()
```

### 14)Analyze FORD's year-wise sales

```python
df['Model Year'] = pd.to_datetime(df['Model Year'], format='%Y')

ford_df = df[df['Make'].str.upper() == 'FORD']

ford_sales_by_year = ford_df['Model Year'].dt.year.value_counts().sort_index()
print(ford_sales_by_year)
plt.figure(figsize=(10, 6))
ford_sales_by_year.plot(kind='line', color='green', marker='o', linestyle='-', linewidth=2, markersize=6)
plt.title('FORD Year-wise Vehicle Sales')
plt.xlabel('Year')
plt.ylabel('Number of Vehicles Sold')
plt.grid(True)
plt.xticks(ford_sales_by_year.index, rotation=45)
plt.tight_layout()
plt.show()
```
